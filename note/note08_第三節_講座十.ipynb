## Splitting with many Keys

```python
 import pandas as pd
```

---
## 用Country切割資料

```python
# summer.columns
split1 = summer.groupby("Country")
```

```python
l = list(split1)
l
```

```python
len(l)
```

```python
# l[100][0]
l[100][1]
# 注意到index有照順序排列!
```

---
## 用Country, Gender來切割資料

```python
split2 = summer.groupby(by = ["Country", "Gender"])
```

```python
l2 = list(split2)
l2
```

```python
len(l2)
```

如果只照國家split，有147個國家  
但是按照國家與性別split，有236個group objects  
所以說明有些國家派出的運動員只有單一性別。

```python
len(l2)
```

```python
l2[104]
```

```python
l2[104][0]
```

```python
l2[104][1]
```

記得，左側的index是和原始數據相同的index
